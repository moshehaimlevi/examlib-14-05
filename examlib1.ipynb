```python
#################### QUESTION NUMBER 2 ####################
import numpy as np
from IPython.display import display
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import drive
drive.mount('/content/drive')
file_path = '/content/drive/My Drive/order_item.csv'
order_item_df = pd.read_csv(file_path)
display(order_item_df)

drive.mount('/content/drive')
file_path = '/content/drive/My Drive/customers.csv'
customers_df = pd.read_csv(file_path)

drive.mount('/content/drive')
file_path = '/content/drive/My Drive/orders.csv'
orders_df = pd.read_csv(file_path)

drive.mount('/content/drive')
file_path = '/content/drive/My Drive/Items.csv'
items_df = pd.read_csv(file_path)
display(items_df)
```

```python
#################### QUESTION A ####################
display(items_df[['item_price']].mean())
```

```python
#################### QUESTION B ####################
merged_df = pd.merge(customers_df, orders_df, left_on='id', right_on='customer_id', how='left')

purchase_count = merged_df.groupby('first_name')['order_id'].count().reset_index()

max_purchases_customer = purchase_count.loc[purchase_count['order_id'].idxmax()]

print(f"Customer with the most purchases: {max_purchases_customer['first_name']}")
print(f"Total number of purchases: {max_purchases_customer['order_id']}")


```

```python
#################### QUESTION GRAPH 1 ####################

plt.figure(figsize=(8, 5))
sns.histplot(data=customers_df, x='gender', bins=30, kde=True, color='green')
plt.title('genders of different customers')
plt.xlabel('genders')
plt.ylabel('customers')
plt.tight_layout()
plt.show()

```

```python
#################### QUESTION GRAPH 2 ####################
plt.figure(figsize=(14, 6))
sns.histplot(data=customers_df, x='nationallity', bins=30, kde=True, color='blue')
plt.title('customers nationality')
plt.xlabel('nationalities')
plt.ylabel('customers')
plt.tight_layout()
plt.show()
```
